```python
from datasets import load_dataset,Dataset
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow import convert_to_tensor, int64
```

```python
tokenizer = Tokenizer()
```

```python
def create_tokenizer(dataset):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(dataset)
    return tokenizer

def find_seq_length(dataset):
    return max(len(seq.split()) for seq in dataset)

def find_vocab_size(tokenizer, dataset):
    tokenizer.fit_on_texts(dataset)
    return len(tokenizer.word_index) + 1
```

```python

train = Dataset.from_csv('train.csv')

# Extract input and target columns 

def make_list(dataset):
    return f""" Movie_name : {dataset['title']}, Length : {dataset['length']}, Release_year : {dataset['release_year']}, Genere : {dataset['genre']}, Plot_summary : {dataset['plot_summary']}, Cast : {dataset['cast']}, imdb_rating : {dataset['rating_imdb']}, Rating_rotten_tomatoes : {dataset['rating_rotten_tomatoes']}"""

def make_promt(train):
    inputs = {}
    train = make_list(train)
    enc_tokenizer = create_tokenizer([inp for inp in train])
    enc_seq_length = find_seq_length([inp for inp in train])
    enc_vocab_size = find_vocab_size(enc_tokenizer, [inp for inp in train])
    trainX = enc_tokenizer.texts_to_sequences([inp for inp in train])
    trainX = pad_sequences(trainX, maxlen=enc_seq_length, padding='post')
    trainX = convert_to_tensor(trainX, dtype=int64)
    inputs['text'] = train
    inputs['input_ids'] = trainX
    print(type(inputs))
    return inputs,enc_tokenizer,enc_vocab_size,enc_seq_length

# train = train.map(make_list)
```

```python
ds = train.map(make_promt)
```
